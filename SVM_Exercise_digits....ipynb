```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
digits = load_digits()
```

```python
digits # cool
```

```python
dir(digits)
```

```python
digits.feature_names
```

Creating a Dataframe to make this more sensible🙃🙃🙃

```python
# using the feature names as the columns
df = pd.DataFrame(data = digits.data, columns = digits.feature_names)
df.head()
```

```python
# adding the target variable..,I mean I need to know what all this numerical something means😏
df['Target'] = digits.target
df
```

```python
digits.target_names # still same.., no point baby🤸‍♀️
```

```python
# I dun even know what I'll plot..,because how many features🤦‍♂️🤦‍♂️🤦‍♂️
plt.scatter(df.pixel_0_0, df.pixel_0_1, color = 'green') # okay...be like say I go add more
plt.scatter(df.pixel_0_2, df.pixel_0_3, color = 'red') # thez one eez nor visualization o, this is confusion
```

```python
# Creating my X and y variables using Feature_names and Target
X = df.drop('Target', axis = 1) # no stress
X # am I actually gonna use all of this🤔🤔
```

```python
y = df['Target']
y
```

```python
# split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Model Time
# Support Vector Classifier this time sha🤺👯‍♂️🤗
model = SVC()
model
```

```python
# Training Time!!!
model.fit(X_train, y_train)
```

```python
# what's the score?👀
model.score(X_test, y_test) # yayyyyyyy!!!(I literally shouted actually(😂🤣🤣)
```

```python
# predict👯‍♂️
model.predict(X_test) # okurrr
```

# Tune the Parameters,baby tune👯‍♂️👯‍♂️

Add params to the model,Fit & Score

Playing around with the Kernel Parameter

```python
model_Ker = SVC(kernel = 'rbf')
model_Ker.fit(X, y)
model_Ker.score(X_test, y_test) # wawwu, I think say nah linear make pass nah🤔
```

```python
model_Ker = SVC(kernel = 'linear')
model_Ker.fit(X, y)
model_Ker.score(X_test, y_test) # drops mic..,and walks away
```

Playing around with the Regularization Parameter

```python
model_Reg = SVC(C = 10)
model_Reg.fit(X, y)
model_Reg.score(X_test, y_test) # okay o...
```

```python
model_Reg = SVC(C = 1)
model_Reg.fit(X, y)
model_Reg.score(X_test, y_test) # well well..,
```

```python
model_Reg = SVC(C = 1000)
model_Reg.fit(X, y)
model_Reg.score(X_test, y_test) # mmhhh
```

Playing around with the Gamma Parameter

```python
model_Gam = SVC(gamma = 1)
model_Gam.fit(X, y)
model_Gam.score(X_test, y_test) # hehehe
```
